```python
# ===== STEP 1: Install required packages =====
!apt-get update -qq
!apt-get install -y -qq tesseract-ocr tesseract-ocr-hin poppler-utils
!pip install -q pytesseract pdf2image pillow

# ===== STEP 2: Imports =====
import pytesseract
from pdf2image import convert_from_path
from google.colab import files
import os

# ===== STEP 3: Upload PDF =====
print("📤 Upload your scanned PDF")
uploaded = files.upload()

pdf_path = list(uploaded.keys())[0]
print(f"✅ Uploaded: {pdf_path}")

# ===== STEP 4: Generate output TXT filename (same as PDF) =====
base_name = os.path.splitext(pdf_path)[0]
output_file = base_name + ".txt"

# ===== STEP 5: Convert PDF pages to images =====
print("🖼 Converting PDF pages to images...")
pages = convert_from_path(pdf_path, dpi=300)

# ===== STEP 6: Perform OCR =====
print("🔍 Performing OCR (Hindi + English)...")
full_text = ""

for i, page in enumerate(pages, start=1):
    print(f"   Processing page {i}/{len(pages)}")
    text = pytesseract.image_to_string(page, lang="hin+eng")
    full_text += f"\n\n----- Page {i} -----\n\n{text}"

# ===== STEP 7: Save OCR output =====
with open(output_file, "w", encoding="utf-8") as f:
    f.write(full_text)

print("✅ OCR completed successfully!")
print(f"📄 TXT saved as: {output_file}")

# ===== STEP 8: Download the TXT file =====
files.download(output_file)
```
